```
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```
def weightsInitialization(input_size,hidden_size,hidden_2,output_size):
    w1 = np.random.randn(input_size,hidden_size)*np.sqrt(1./input_size)
    b1 = np.zeros((1,hidden_size))
    w2 = np.random.randn(hidden_size,hidden_2)*np.sqrt(1./hidden_size)
    b2 = np.zeros((1,hidden_2))
    w3 = np.random.randn(hidden_2,output_size)*np.sqrt(1./hidden_2)
    b3 = np.zeros((1,output_size))
    return w1,b1,w2,b2,w3,b3
```

```
def softmax(x):
    shift_x = x - np.max(x, axis=1, keepdims=True)
    exp_x = np.exp(shift_x)
    return exp_x / np.sum(exp_x, axis=1, keepdims=True)
def relu(x):
    return np.maximum(0,x)

def relu_derivative(x):
    return np.where(x>0,1,0)
```

```
def CategoricalCrossEntropy(y,y_pred,w1,w2,w3,lambda_reg):
    epsilon = 1e-12
    y_pred = np.clip(y_pred, epsilon, 1. - epsilon)
    cross_entropy_loss = np.mean(-np.sum(y * np.log(y_pred), axis=1))
    reg_term = (lambda_reg / (2 * y.shape[0])) * (np.sum(np.square(w1)) + np.sum(np.square(w2)) + np.sum(np.square(w3)))
    total_loss = cross_entropy_loss + reg_term
    return total_loss
```

```
def forward_pass(X,y,w1,b1,w2,b2,w3,b3):
    z1= X@w1+b1
    A1 = relu(z1)
    z2 = A1@w2+b2
    A2 = relu(z2)
    z3 = A2@w3+b3
    A3 = softmax(z3)
    return z1,A1,z2,A2,z3,A3
```

```
def backward_pass(X,y,z1,A1,z2,A2,z3,A3,w1,b1,w2,b2,w3,b3,lambda_reg,mini_batch_size):
    m=X.shape[0]
    dz3 = A3 - y
    dw3 = (A2.T @ dz3)/ mini_batch_size + (lambda_reg / mini_batch_size) * w3
    db3 = np.mean(dz3,axis = 0,keepdims=True)

    dA2 = dz3 @ w3.T
    dz2 = dA2 * relu_derivative(A2)
    dw2 = (A1.T @ dz2)/ mini_batch_size + (lambda_reg / mini_batch_size) * w2
    db2 = np.mean(dz2,axis=0,keepdims=True)

    dA1 = dz2 @ w2.T
    dz1 = dA1 * relu_derivative(A1)
    dw1 = (X.T @ dz1)/ mini_batch_size + (lambda_reg / mini_batch_size) * w1
    db1 = np.mean(dz1,axis=0,keepdims=True)

    return dw1,db1,dw2,db2,dw3,db3
```

```
def train_model(X,y,w1,b1,w2,b2,w3,b3,epochs,L,lambda_reg,batch_size):
    losses =[]
    m= X.shape[0]
    for epoch in range(epochs):
        indi = np.random.permutation(X.shape[0])
        X= X[indi]
        y= y[indi]

        num_batches = int(np.ceil(m/batch_size))

        for batch in range(num_batches):
            start_idx = batch*batch_size
            end_idx = min(start_idx+batch_size,m)
            X_batch = X[start_idx:end_idx]
            y_batch = y[start_idx:end_idx]
            mini_batch_size = X_batch.shape[0]
            z1,A1,z2,A2,z3,A3= forward_pass(X_batch,y_batch,w1,b1,w2,b2,w3,b3)
            dw1,db1,dw2,db2,dw3,db3 = backward_pass(X_batch,y_batch,z1,A1,z2,A2,z3,A3,w1,b1,w2,b2,w3,b3,lambda_reg,mini_batch_size)
            w1 -= L * dw1
            b1 -= L * db1
            w2 -= L * dw2
            b2 -= L * db2
            w3 -= L* dw3
            b3 -= L * db3
        z1_epoch ,A1_epoch,z2_epoch,A2_epoch,z3_epoch,A3_epoch = forward_pass(X,y,w1,b1,w2,b2,w3,b3)
        loss = CategoricalCrossEntropy(y,A3_epoch,w1,w2,w3,lambda_reg)
        losses.append(loss)
        if(epoch%1000 ==0):
            print(f"epoch: {epoch} loss : {loss:.4f}")
    return w1,b1,w2,b2,w3,b3,losses
```

```
def predict(w1,b1,w2,b2,w3,b3,X):
    z1= X@w1+b1
    A1 = relu(z1)
    z2 = A1@w2+b2
    A2 = relu(z2)
    z3 = A2@w3+b3
    A3 = softmax(z3)
    return (A3)
```
